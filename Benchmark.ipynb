```python
import matplotlib.pyplot as plt
```

```python
# 100000 repeats of algorithm
length = [1, 5, 10, 19]
time_100000 = [
    [123, 395, 916, 2390],
    [126, 427, 845, 2268],
    [147, 437, 845, 2292]
]

# 50 repeats of algorithm
time_50 = [
    [16, 28, 36, 52],
    [24, 25, 29, 42],
    [20, 20, 24, 37]
]

# Pimitive algorithm
time_primitive = [
    [1, 1, 5, 18076],
    [1, 2, 6, 18762],
    [1, 2, 4, 20436]
]
```

```python
time_avg_100000 = [0, 0, 0, 0]
for tuple_time in time_100000:
    for i in range(4):
        time_avg_100000[i] += tuple_time[i]
time_avg_100000 = [x / 3 for x in time_avg_100000]

time_avg_50 = [0, 0, 0, 0]
for tuple_time in time_50:
    for i in range(4):
        time_avg_50[i] += tuple_time[i]
time_avg_50 = [x / 3 for x in time_avg_50]

time_primitive_avg = [0, 0, 0, 0]
for tuple_time in time_primitive:
    for i in range(4):
        time_primitive_avg[i] += tuple_time[i]
time_primitive_avg = [x / 3 for x in time_primitive_avg]
```

```python
plt.plot(length, time_avg_100000, label="Miller-Rabin 100000 iterations", c="orangered")
plt.plot(length, time_avg_50, label="Miller-Rabin 50 iterations", c="firebrick")
plt.legend()
plt.rcParams["figure.figsize"] = [12, 5]
plt.xlabel('number length')
plt.ylabel('milliseconds')
plt.show()
```

```python
plt.plot(length, time_avg_100000, label="Miller-Rabin 100000 iterations", c="orangered")
plt.plot(length, time_avg_50, label="Miller-Rabin 50 iterations", c="firebrick")
plt.plot(length, time_primitive_avg, label="primitive algorithm", c="royalblue")
plt.legend()
plt.rcParams["figure.figsize"] = [12, 5]
plt.xlabel('number length')
plt.ylabel('milliseconds')
plt.show()
```
